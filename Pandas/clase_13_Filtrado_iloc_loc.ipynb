### Filtrar con loc
Acceda a un grupo de filas y columnas por etiqueta(s) o una matriz booleana.

A diferencia del el slicing normal, que corta por la pos del index con dataframe. loc tenes la posicion de start y stop para filtrar.

#### Filtrar con Slicing

```python

df_books.loc[0:4]
```

#### Filtrar varios tags 

```python
df_books.loc[0:4, ['Name','User Rating','Reviews']]
```

#### Filtrar y modificar

```python
df_books.loc[0:4, ['User Rating']] * 20
```

#### Filtrar por igualdad

```python
# Devuelve una lista booleana de quien cumple la condicion
df_books.loc[:, ['Author']] == 'Jen Sincero	'
```

### Filtrar con iloc
Filtra puramente por la indexación en la ubicación de enteros para la selección por posición.

.iloc[] se basa principalmente en la posición de números enteros (de 0 a la longitud 1 del eje), pero también se puede usar con una matriz booleana.

![](./data/loc.PNG)

#### Filtrar con Slicing

```python
df_books.iloc[:]
```

#### Filtrar por indice

![](./data/Captura.PNG)

```python
df_books.iloc[:,0:3]
```

```python
df_books.iloc[1,3] * -1
```

```python
df_books.iloc[:2,2:]
```
