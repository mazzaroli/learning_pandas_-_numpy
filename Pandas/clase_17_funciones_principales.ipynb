# Funciones Principales

## .info()
Imprime un resumen conciso de un DataFrame.

Este método imprime información sobre un DataFrame, incluido el tipo de índice y las columnas, los valores no nulos y el uso de la memoria.

```python
df_books.info()
```

## describe()
Genera estadísticas descriptivas.

Las estadísticas descriptivas incluyen aquellas que resumen la tendencia central, la dispersión y la forma de la distribución de un conjunto de datos, excluyendo NaNlos valores.

Analiza series numéricas y de objetos, así como DataFrameconjuntos de columnas de tipos de datos mixtos. La salida variará dependiendo de lo que se proporcione. Consulte las notas a continuación para obtener más detalles.

```python
df_books.describe()
```

## tail()
Devuelve las últimas n filas.

Esta función devuelve las últimas n filas del objeto según la posición. Es útil para verificar datos rápidamente, por ejemplo, después de ordenar o agregar filas.

Para valores negativos de n , esta función devuelve todas las filas excepto las primeras n filas, equivalente a df[n:].

```python
df_books.tail(2)
```

## memory_usage()
Devuelve el uso de memoria de cada columna en bytes. 

El uso de la memoria puede incluir opcionalmente la contribución del índice y los elementos del objeto dtype. 



```python
df_books.memory_usage(deep=True)
```

```python
df_books.memory_usage()
```

## value_counts()
Devuelve una Serie que contiene recuentos de filas únicas en el DataFrame.


```python
df_books["Author"].value_counts()
```

## drop_duplicates()
Devuelve DataFrame con filas duplicadas eliminadas.

Considerar ciertas columnas es opcional. Los índices, incluidos los índices de tiempo, se ignoran.

```python
df_books.iloc[0]
```

```python
df_books = df_books.append(df_books.iloc[0])
```

```python
df_books.drop_duplicates(keep='last')
```

## sort_values()
*DataFrame.sort_values(by, axis=0, ascending=True, inplace=False, kind='quicksort', na_position='last', ignore_index=False, key=None)*

Ordene por los valores a lo largo de cualquier eje.

```python
df_books.sort_values('Year', ascending=False)
```
