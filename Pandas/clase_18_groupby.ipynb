```python
import pandas as pd
```

```python
df_books = pd.read_csv('./data/bestsellers-with-categories.csv', sep=',', header=0)
df_books.head(1)
```

# Groupby
[Documentacion](https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.groupby.html?highlight=groupby#pandas.DataFrame.groupby)

## Agrupar por un Tag

```python
df_books.groupby(by="Author").count().head(5)
```

## Agrupar con operaciones arimeticas

```python
df_books.groupby(by="Author").mean().head(5)
```

```python
df_books.groupby(by="Author").min().head(5)
```

```python
df_books.groupby(by="Author").max().head(5)
```

```python
df_books.groupby(by='Author').sum().head(5)
```

## Agrupar filtrando por Indice

```python
df_books.groupby(by='Author').sum().loc['Adir Levy']
```

## Agrupar resetear indice

```python
df_books.groupby(by='Author').sum().reset_index().loc[1]
```

## Agrupar por funciones de agregacion
Para aplicar una sola función al agrupamiento tambien se puede usar la función .apply().

La función .agg() se utiliza cuando se desean hacer dos o más cálculos.

```python
df_books.groupby('Author').agg(['min','max'])
```

## Agrupar datos especificos de X cantidad de Tags

```python
df_books.groupby('Author').agg({'Reviews':['min','max'], 'User Rating':'sum'})
```

## Agrupar multiples condiciones por llave de agrupamiento

```python
df_books.groupby(['Author','Year']).count()
```
